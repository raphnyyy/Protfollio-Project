```python
#Adjust the data range to [0,1] with normalization.
#Define Normalization formula
def transform_with_normalization(vals):
    new_vals = []
    max_val = max(vals)
    min_val = min(vals)
    for val in vals:
        val_norm = (val-min_val)/(max_val-min_val)
        new_vals.append(val_norm)
    return new_vals
```

```python
#Create new columns of each original columns for data that has a range difference using normalization.
for col in new_apple_value_cloumns:
    df_apple_value[col + " Norm"] = transform_with_normalization(df_apple_value[col].values)
```
